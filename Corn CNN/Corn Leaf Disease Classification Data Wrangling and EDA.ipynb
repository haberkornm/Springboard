# Loading Data

```python
from tensorflow import keras
import pathlib
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
#Loading data into keras
data_dir = 'data'
dataset = tf.keras.preprocessing.image_dataset_from_directory(data_dir, 
                                                              image_size=(180, 180), 
                                                              batch_size=64, 
                                                              labels='inferred',  
                                                              label_mode='categorical')
```

```python
#Listing directories in folder
os.listdir(data_dir)
```

```python
#counting number of images in each directory
Blight = glob(os.path.join(data_dir+"/Blight/"+"*"))
Common_Rust = glob(os.path.join(data_dir+"/Common_Rust/"+"*"))
Gray_Leaf_Spot = glob(os.path.join(data_dir+"/Gray_Leaf_Spot/"+"*"))
healthy = glob(os.path.join(data_dir+"/Healthy/"+"*"))
print('Blight images:', len(Blight))
print('Rust images:', len(Common_Rust))
print('Spot images:', len(Gray_Leaf_Spot))
print('Healthy images:', len(healthy))
print('Total images:', len(Blight)+len(Common_Rust)+len(Gray_Leaf_Spot)+len(healthy))
```

```python
#Creating labels for images
import pandas as pd
labels = pd.DataFrame(['Blight'] * len(Blight) + ['Common_Rust'] * len(Common_Rust) + ['Gray_Leaf_Spot'] * len(Gray_Leaf_Spot) + ['Healthy'] * len(healthy))
print(labels.value_counts())
print('total labels:', len(labels))
```

# Viewing Images

```python
#Viewing blight corn leaves
plt.suptitle('Blight', fontsize=16)
for i in list(range(4)):
    plt.subplot(2,2,i+1)
    plt.imshow(plt.imread(Blight[i]))
```

```python
#viewing gray leaf spot leaves
plt.suptitle('Gray Leaf Spot', fontsize=16)
for i in list(range(4)):
    plt.subplot(2,2,i+1)
    plt.imshow(plt.imread(Gray_Leaf_Spot[i]))
```

```python
#viewing healthy corn leaves
plt.suptitle('Healthy', fontsize=16)
for i in list(range(4)):
    plt.subplot(2,2,i+1)
    plt.imshow(plt.imread(healthy[i]))
```
